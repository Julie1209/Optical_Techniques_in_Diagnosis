```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_set = {}
```

```python
# 產生 5 筆 random_set，每筆有 10000 個 data
for i in range(5):
    random_set[str(i)] = np.random.uniform(low=0, high=1, size=10000)
```

```python
# (A) Determine the frequency of occurrence of the random numbers in each interval 
# and display the result as a bar graph (this is called a histogram).
```

```python
# random_set 1 histogram
plt.figure(figsize=(8,6))
counts_0, bins_0, patches_0 = plt.hist(random_set["0"], bins=20, range=(0, 1), rwidth=0.9)
plt.xlabel("the region [0, 1]", fontsize=15)
plt.ylabel("frequence of occurence", fontsize=15)
plt.title("Histogram of random_set 1", fontsize=15)
plt.savefig("Histogram of random_set 1.png")
plt.show()
print("Frequency of occurrence of the random numbers in each interval of random_set 1:\n{}\n".format(counts_0))
```

```python
# random_set 2 histogram
plt.figure(figsize=(8,6))
counts_1, bins_1, patches_1 = plt.hist(random_set["1"], bins=20, range=(0, 1), rwidth=0.9)
plt.xlabel("the region [0, 1]", fontsize=15)
plt.ylabel("frequence of occurence", fontsize=15)
plt.title("Histogram of random_set 2", fontsize=15)
plt.savefig("Histogram of random_set 2.png")
plt.show()
print("Frequency of occurrence of the random numbers in each interval of random_set 2:\n{}\n".format(counts_1))
```

```python
# random_set 3 histogram
plt.figure(figsize=(8,6))
counts_2, bins_2, patches_2 = plt.hist(random_set["2"], bins=20, range=(0, 1), rwidth=0.9)
plt.xlabel("the region [0, 1]", fontsize=15)
plt.ylabel("frequence of occurence", fontsize=15)
plt.title("Histogram of random_set 3", fontsize=15)
plt.savefig("Histogram of random_set 3.png")
plt.show()
print("Frequency of occurrence of the random numbers in each interval of random_set 3:\n{}\n".format(counts_2))
```

```python
# random_set 4 histogram
plt.figure(figsize=(8,6))
counts_3, bins_3, patches_3 = plt.hist(random_set["3"], bins=20, range=(0, 1), rwidth=0.9)
plt.xlabel("the region [0, 1]", fontsize=15)
plt.ylabel("frequence of occurence", fontsize=15)
plt.title("Histogram of random_set 4", fontsize=15)
plt.savefig("Histogram of random_set 4.png")
plt.show()
print("Frequency of occurrence of the random numbers in each interval of random_set 4:\n{}\n".format(counts_3))
```

```python
# random_set 5 histogram
plt.figure(figsize=(8,6))
counts_4, bins_4, patches_4 = plt.hist(random_set["4"], bins=20, range=(0, 1), rwidth=0.9)
plt.xlabel("the region [0, 1]", fontsize=15)
plt.ylabel("frequence of occurence", fontsize=15)
plt.title("Histogram of random_set 5", fontsize=15)
plt.savefig("Histogram of random_set 5.png")
plt.show()
print("Frequency of occurrence of the random numbers in each interval of random_set 5:\n{}\n".format(counts_4))
```

```python
# (B) For each of the 20 intervals, determine the mean and standard deviation over the five runs. 
# Ideally the mean of each interval is about 10,000/20 = 500 and the standard deviation should be less than 36.
```

```python
# 合併 5 個 random_set 的 frequency of occurrence of the random numbers in each interval
counts_set = np.stack((counts_0,                             
                       counts_1,                             
                       counts_2,                             
                       counts_3,                             
                       counts_4), axis=0)
print("合併後資料(frequency of occurrence of the random numbers in each interval over the five runs):\n{}\n".format(counts_set))
print("形狀:\n{}\n".format(counts_set.shape))
```

```python
# 重排資料，以各個 run 的第 1, 第 2, ...個 interval 的 frequency of occurrence of the random numbers 為順序
counts_set = counts_set.T
print("重排後的資料:\n{}\n".format(counts_set))
print("形狀:\n{}\n".format(counts_set.shape))
```

```python
Q2_result = {str(i):{"mean":None, "std":None} for i in range(20)}
```

```python
# 計算 mean and standard deviation over the five runs.(for each interval)
for i in range(20):
    Q2_result[str(i)]["mean"] = np.mean(counts_set[i])
    Q2_result[str(i)]["std"] = np.std(counts_set[i], ddof=1)
print("每一個 interval 的平均數和標準差(over the five runs):\n")

for i in range(20):
    print("第 {:2} 個 interval: ".format(i), end="")
    print(Q2_result[str(i)])
```
